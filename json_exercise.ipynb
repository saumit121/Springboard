## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

## JSON exercise

Using data in file 'data/world_bank_projects.json' and the techniques demonstrated above,
1. Find the 10 countries with most projects
2. Find the top 10 major project themes (using column 'mjtheme_namecode')
3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
#load in world bank data as string and as data frame
pd.set_option('display.max_columns', 500)

w = json.load((open('data/world_bank_projects.json')))
df = pd.read_json('data/world_bank_projects.json')
df.head()
```

```python
df.info()
```

1) Find the top 10 countries with the most projects.

```python
pd.set_option('display.max_rows', 500)
df.groupby('countryshortname')['countryshortname'].count().sort_values(ascending = False).head(10)
```

2) Find the top 10 major project themes (using column 'mjtheme_namecode')

```python
#Inefficient way unpacking each row of columns to get project theme name
theme_list = []
for i in df['mjtheme_namecode']:
    for dic in i:
        for k,v in dic.items():
            if k == 'name':
                theme_list.append(v)
```

```python
#Convert list to Pandas series
theme_df = pd.Series(theme_list)
```

```python
theme_df.value_counts()
```

```python
#Use normalization method instead

theme = json_normalize(w, 'mjtheme_namecode')['name']
theme = pd.Series(theme)
theme.value_counts()
```

3) In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

There are 122 entries with missing project themes. 2 approaches will be used to impute the missing values. The first is to simply replace the empty entries with 'Missing Name'. The second approach is to match the name of an entry with the same code as that of the missing entry. 

```python
theme = pd.DataFrame(theme)
```

```python
#Replace missing entries
theme['name'] = theme['name'].replace('','Missing Name')
```

```python
#Verify all blank entries have been replaced with 'Missing Name'
theme['name'].value_counts()
```

```python
#Normalize again

theme_norm = json_normalize(w, 'mjtheme_namecode')
theme_norm = pd.DataFrame(theme_norm)
theme_norm['code'] = theme_norm['code'].astype('int64')

theme_norm['name'].unique()
```

```python
#Create mapping dictionary with code/name combinations

codename_dict = {1: 'Economic management', 2: 'Public sector governance', 3: 'Rule of law', 4: 'Financial and private sector development',
                5: 'Trade and integration', 6: 'Social protection and risk management', 7: 'Social dev/gender/inclusion',
                8: 'Human development', 9: 'Urban development', 10: 'Rural development', 11: 'Environment and natural resources management'}

codename_dict
```

```python
theme_norm.drop(['name'], axis = 1, inplace = True)
theme_norm
```

```python
#Replace missing name values with mapped dictionary values
theme_norm['name'] = theme_norm['code'].replace(codename_dict)
```

```python
#There are no longer any missing/blank values
theme_norm['name'].value_counts()
```
